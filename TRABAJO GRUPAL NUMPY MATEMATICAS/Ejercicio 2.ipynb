```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
def modelo(y, t):
    dydt = -y + t**2 + 1
    return dydt
```

```python
y0 = 0
t = np.linspace(0, 5, 100)
```

```python
sol = odeint(modelo, y0, t)
```

```python
plt.plot(t, sol)
plt.xlabel('t')
plt.ylabel('y(t)')
plt.title('Solución de la EDO dy/dt = -y + t^2 + 1')
plt.grid()
plt.show()
```
