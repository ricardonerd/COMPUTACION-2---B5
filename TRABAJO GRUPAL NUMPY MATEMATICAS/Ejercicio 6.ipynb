```python
# Para gráfica de f(x) = sin(x), x ∈ [-2, 2]

# Definición del intervalo
x1 = np.linspace(-2, 2, 400)

# Definición de la función
f1 = np.sin(x1)

# Graficar la función
plt.figure(figsize=(7, 5))
plt.plot(x1, f1, label=r'$f(x) = \sin(x)$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title(r'$f(x) = \sin(x)$, $x \in [-2, 2]$')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Para gráfica de f(x) = e^(3x), x ∈ [-2, 4]
# Definición del intervalo
x2 = np.linspace(-2, 4, 400)

# Definición de la función
f2 = np.exp(3 * x2)

# Graficar la función
plt.figure(figsize=(7, 5))
plt.plot(x2, f2, label=r'$f(x) = e^{3x}$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title(r'$f(x) = e^{3x}$, $x \in [-2, 4]$')
plt.grid(True)
plt.legend()
plt.show()
```
