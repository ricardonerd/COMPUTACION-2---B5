```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definir la función f(x)
def f(x):
    return np.exp(-x**2)
```

```python
# Definir el dominio
x = np.linspace(-1, 3, 400)
```

```python
# Calcular los valores de la función
y = f(x)
```

```python
# Crear la gráfica
plt.plot(x, y, label='Función f(x)')
plt.xlabel('Eje x')
plt.ylabel('f(x)')
plt.title('Función f(x)')
plt.legend()
plt.grid(True)
plt.show()
```
