```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definir la función f(x)
def f(x):
    return np.exp(-x**2)
```

```python
# Definir la función 1 - f(x)
def g(x):
    return 1 - f(x)
```

```python
# Definir el dominio
x = np.linspace(-1, 3, 400)
```

```python
# Calcular los valores de las funciones
y_f = f(x)
y_g = g(x)
```

```python

# Crear la gráfica
plt.plot(x, y_f, 'ro', label='Función f(x)')  # Línea roja con círculos
plt.plot(x, y_g, 'g--', label='Función 1 - f(x)')  # Línea verde discontinua
plt.xlabel('Eje x')
plt.ylabel('f(x) y 1 - f(x)')
plt.title('Funciones f(x) y 1 - f(x)')
plt.legend()
plt.grid(True)
plt.show()
```
