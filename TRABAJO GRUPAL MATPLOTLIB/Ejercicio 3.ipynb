```python
Ejercicio 3: La función scatter muestra una nube de puntos, con posibilidad de variar también el tamaño y el color.
llegando a gráficas asi:
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos aleatorios
np.random.seed(42)
x = np.random.randn(100)
y = np.random.randn(100)
colors = np.random.randn(100)  # Generar colores aleatorios
sizes = 100 * np.random.rand(100)

# Características de la gráfica
plt.figure(figsize=(8, 6))
scatter = plt.scatter(x, y, s=sizes, c=colors, cmap="Blues", alpha=0.6, edgecolors="black", linewidth=0.5)
plt.colorbar(scatter)

# Definir título 
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.title("Gráfica de dispersión")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos aleatorios
np.random.seed(42)
x = np.random.randn(100)
y = np.random.randn(100)
colors = np.random.randn(100)  # Generar colores aleatorios
sizes = 100 * np.random.rand(100)

# Características de la gráfica
plt.figure(figsize=(8, 6))
scatter = plt.scatter(x, y, s=sizes, c=colors, cmap="Oranges", alpha=0.6, edgecolors="black", linewidth=0.5)
plt.colorbar(scatter)

# Definir título 
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.title("Gráfica de dispersión")

plt.show()
```

```python
Con parametros s y c podemos modificar el tamaño y el color respectivamente. Para el color, a cada valor numérico se le asigna un color a través de un mapa de colores; ese mapa se puede cambiar con el argumento cmap. Esa correspondencia se puede visualizar llamando a la función colorbar.
```
